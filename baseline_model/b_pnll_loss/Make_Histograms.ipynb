```python
import numpy as np
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("likelihoods.csv")
data.head()
```

```python
null = list(data["lhs_null"])
model = list(data["lhs_model"])
saturated = list(data["lhs_saturated"])
```

```python
num_bins = 10

n, bins, patches = plt.hist(null, num_bins, facecolor='blue', alpha=0.5, label="null")
n, bins, patches = plt.hist(model, num_bins, facecolor='yellow', alpha=0.5, label="baseline")
n, bins, patches = plt.hist(saturated, num_bins, facecolor='green', alpha=0.5, label="saturated")
plt.legend()
plt.title("Likelihoods per Image")
plt.ylabel("#Images")
plt.xlabel("Likelihood")
plt.show()
```
